```python
import math
import random
import sys


def MAE(lst_target, lst_predict):
    n = len(lst_predict)
    return 1/n * sum(math.fabs(lst_target[i] - lst_predict[i]) for i in range(n))

def MSE (lst_target ,lst_predict):
    n = len(lst_predict)
    return 1/n * sum(  (lst_target[i] - lst_predict[i])**2 for i in range (n))

def RMSE ( lst_target , lst_predict ):
    return math.sqrt(MSE(lst_target , lst_predict))

def Huber_Loss ( lst_target , lst_predict ):
    n = len(lst_predict)
    alpha = 0.5
    sumloss = 0.0
    for i in range (n):
        if math.fabs(lst_target[i] - lst_predict[i] ) <= alpha:
            sumloss += 0.5 *( lst_target[i] - lst_predict[i] )**2
        else :
            sumloss += alpha*(math.fabs(lst_target[i] - lst_predict[i]) - 0.5*alpha )
    return sumloss


def enter_number ( ):
    number_sample=0
    try :
        number_sample = int(input(">>Input number of samples (positive integer number) which are generated :"))
    except Exception :
        print("number of samples must be a positive integer number")
        sys.exit()

    target = [ random.uniform(0,10) for i in range (11)]
    prediction = [random.uniform(0,10)  for i in range (11)]
    return target,prediction
def check_name ():
    name = input("Input loss name (MAE, MSE, RMSE, Huber_Loss): ")
    arr = ['MAE', 'MSE', 'RMSE','Huber_Loss' ]
    if name not in arr :
        print("loss name loss is not supported")
        sys.exit()
    return name
# -- main ----
target, prediction =enter_number()
name = check_name()
print("exercise 1 :")

print("Target: ",target)
print("Prediction: ",prediction)

if name == "MAE":
    print("MAE: ", MAE(target,prediction))
elif name == "MSE":
    print("MSE: ", MSE(target,prediction))
elif name == "RMSE":
    print("RMSE: ", RMSE(target, prediction))
elif name == "Huber loss":
    print("Huber loss:", Huber_Loss(target,prediction))
```
