```python
#bai tap ve tuple
import math

n = int(input("nhap so luong phan tu tuple 1"))
tupl1 = tuple( int(input()) for i in range (n))



aver = sum(tupl1) / len(tupl1)
moreAverage =0
moreAverage  = sum( 1 for i in tupl1 if i > aver )
print("so phan tu lon hon gia tri trung binh la:", moreAverage)

tupl1_odd = tuple( i for i in tupl1 if i %2 ==1 )
tupl1_even = tuple( i for i in tupl1 if i %2 ==0 )
print("Tuple các số lẻ:", tupl1_odd)
print("Tuple các số chẵn:", tupl1_even)

n2= int(input("nhap so phan tu tuple 2"))
tupl2 = tuple( input() for i in range(n2))

k = input("nhap ki tu muon tim kiem") 
g = sum ( 1 for j in tupl2 for i in j if k==i )
print(f"Số lần ký tự '{g}' xuất hiện trong tuple 2 là:", k)


print([ i for i in tupl2 if len(i) >= 5 ])


# ghep a và b để được tuple c

minlen = min( len ( tupl1) , len ( tupl2 ))

tupl3 = tuple( str(tupl1[i]) + tupl2[i] for i in range(minlen) ) + tupl1[minlen:] + tupl2[minlen:]

print(tupl1)
print(tupl2)
print(tupl3)
 
```

```python
#quan ly tai khoan sinh vien clb
from random import *
import pprint

n = int(input("nhap vao so account: "))
list_pass = ['CNTT', 'KHMT','KTPM', 'HTTT', 'DAPT']
user_name = 2023600000

danhsach ={}
for i in range (n):
    user_name+=1
    password = choice(list_pass)+str(user_name)
    danhsach [f"Account{i+1}"] = {"userName" : user_name , "password" : password}

pprint.pprint(danhsach)

```

```python
# file docker-compose 
import pprint
docker_compose = {
'version' : '3.8',
'services' :  'app',
'image' : 'python:3.10-slim',
'command' : 'python app.py',
'volumes' :'./app:/app',
'restart': 'always',
'ports': '5000:5000',
'depends_on' : 'db'}

pprint.pprint(docker_compose)

# sua version
docker_compose['version'] = '3.10'
print('\nSau khi thay doi version:\n ')
pprint.pprint(docker_compose)

#in gia tri cua image
print("\nGia tri cua image: \n")
print( docker_compose["image"])

#them environment
docker_compose["environment"] = 'PYTHONUNBUFFERED=1'
print('\nSau khi thay doi :\n ')
pprint.pprint(docker_compose)

#kiem tra ton tai cua volumes
if 'volumes' in docker_compose:
    print("'volumes' có trong dictionary.")
else:
    print("'volumes' không có trong dictionary.")

# xoa depend_on
del docker_compose['depends_on']
#in ra so luong key
print('so luong key: ' , len(docker_compose))

#tao list chua cac value
my_list = [ value for value in docker_compose.values()]
print(my_list)

# alway
flag = False
for i in docker_compose.value:
    if i == 'alway':
        print("Co alway trong value cua dict.")
        flag = True
        break
if flag == False :
    print("Ko tim thay alway trong value cua dict")

# nhapvao 1 cap key value va them vao dict
docker_compose[input("nhap vao key: ")] = input("nhap vao value: ")
print('\nSau khi thay doi :\n ')
pprint.pprint(docker_compose)
```
